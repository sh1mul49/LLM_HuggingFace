```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, TrainingArguments, Trainer, DataCollatorForLanguageModeling

#model_name = "bigcode/starcoderbase-1b"
model_name = "deepseek-ai/deepseek-coder-1.3b-base"

# Modified quant_type to "nf4" for CPU compatibility
bnb_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_use_double_quant=True, bnb_4bit_compute_dtype="float16", bnb_4bit_quant_type="nf4")

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")
```

```python
import json

# Load JSON file
with open('/content/drive/MyDrive/Bulipe/TicTacToe.json', 'r') as f:
    data = json.load(f)

# Access data
print(data)
```

```python
def tokenize(sample):
    return tokenizer(sample["text"] + "\n" + sample["code"], truncation=True, padding="max_length", max_length=2048)

from datasets import Dataset

# Convert the list to a Dataset object
dataset = Dataset.from_list(data)

tokenized_dataset = dataset.map(tokenize, remove_columns=["text", "code"])
```

```python
from peft import get_peft_model, LoraConfig, TaskType

peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=16,
    lora_alpha=64,
    lora_dropout=0.05
)

# Apply the PEFT model to the base model
model = get_peft_model(model, peft_config)

# Print the number of trainable parameters to confirm PEFT is applied
model.print_trainable_parameters()
```

```python
training_args = TrainingArguments(
    output_dir="./GameTicTacToe_model",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=2,
    save_strategy="epoch",  # keep this if supported
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    logging_dir="./logs",
    save_total_limit=2
)

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

trainer = Trainer(
    model=model,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer,
    args=training_args,
    data_collator=data_collator
)
```

```python
training_args.report_to = []
```

```python
import transformers

for callback in trainer.callback_handler.callbacks:
    if isinstance(callback, transformers.integrations.integration_utils.WandbCallback):
        trainer.callback_handler.remove_callback(callback)
        break
```

```python
trainer.train()
```

```python
from transformers import pipeline, AutoTokenizer
import webbrowser

# Load your trained model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("./Game763_model")
generator = pipeline("text-generation", model="./Game763_model", tokenizer=tokenizer, device=0)

# Set the game prompt — you can modify the game type (e.g., Snake, Tetris, 2048, etc.)
game_type = "Tetris"  # Change to: "Snake", "Breakout", "Memory", "2048", "Space Invaders", etc.

prompt = f"""
### Instruction:
Generate HTML/JavaScript {game_type} Game:
Create a playable mini {game_type.lower()} game using HTML, CSS and JavaScript.
Include game controls and collision detection if applicable.

### Code:
"""

# Generate code using the model
output = generator(
    prompt,
    max_new_tokens=2500,
    temperature=0.3,
    top_p=0.95,
    num_return_sequences=1
)[0]["generated_text"]

# Extract the code portion
generated_code = output.split("### Code:")[1].strip()

# Save the generated code to an HTML file
filename = f"generated_{game_type.lower()}_game.html"
with open(filename, "w", encoding="utf-8") as f:
    f.write(generated_code)

# Open the generated HTML file in the default web browser
webbrowser.open(filename)

# Optional: print confirmation
print(f"\n✅ {game_type} game code generated and saved to '{filename}'. It should open in your browser now.")
```
